```python
import pandas as pd
import numpy as np
from quantfreedom.evaluators import is_below, combine_evals, is_above
from quantfreedom.indicators.talib_ind import from_talib
price_data = pd.read_hdf('../data/prices.hd5')
rsi_ind = from_talib(
    func_name='rsi',
    price_data=price_data,
    cart_product=False,
    combos=False,
    timeperiod=[15, 30],
)
ema_ind = from_talib(
    func_name='ema',
    price_data=price_data,
    cart_product=False,
    combos=False,
    timeperiod=[100, 200, 300],
)
ema_2_ind = from_talib(
    func_name='ema',
    price_data=price_data,
    cart_product=False,
    combos=False,
    timeperiod=[600, 800, 1000],
)
rsi_ema_ind = from_talib(
    func_name='ema',
    indicator_data=rsi_ind,
    cart_product=False,
    combos=False,
    timeperiod=[100, 200, 300],
)
ema_is_above = is_above(
    want_to_evaluate=ema_ind,
    price_data=price_data,
    candle_ohlc='close'
)
ema_is_below = is_below(
    want_to_evaluate=ema_ind,
    price_data=price_data,
    candle_ohlc='close'
)
ema_2_is_above = is_above(
    want_to_evaluate=ema_2_ind,
    price_data=price_data,
    candle_ohlc='close'
)
ema_2_is_below = is_below(
    want_to_evaluate=ema_2_ind,
    price_data=price_data,
    candle_ohlc='close'
)
rsi_is_below = is_below(
    want_to_evaluate=rsi_ind,
    user_args=[50,60],
)
rsi_ema_is_below = is_below(
    want_to_evaluate=rsi_ema_ind,
    indicator_data=rsi_ind,
)
```
