```python
import numpy as np
import pandas as pd
```

```python
array = np.random.randint(1,20,(6,4))
array
```

```python
pd.DataFrame(array)
```

```python
df_columns = pd.MultiIndex.from_tuples([(20, 300), (20, 600),(50,300), (50,600)], names=['RSI Below', 'Price Above EMA'])
df_columns
```

```python
pd.DataFrame(array, columns=df_columns)
```

```python
df_index = pd.MultiIndex.from_tuples([(2017, 4), (2017, 8),(2017,12), (2018,4), (2018,8), (2018,12)], names=['Year', 'Month'])
df_index
```

```python
pd.DataFrame(array, index=df_index, columns=df_columns)
```

```python
df = pd.DataFrame(array, index=df_index, columns=df_columns)
df
```

```python
df.xs(600, level='Price Above EMA', axis=1)
```

```python
df.xs(300, level=1, axis=1)
```

```python
df.xs(20, level=0, axis=1)
```
