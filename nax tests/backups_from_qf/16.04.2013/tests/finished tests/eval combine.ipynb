```python
import pandas as pd
from quantfreedom.evaluators import is_below, combine_evals, is_above
from quantfreedom.indicators.talib_ind import from_talib
prices = pd.read_hdf('../data/prices.hd5')
```

```python
rsi_ind = from_talib(
    func_name='rsi',
    price_data=prices,
    cart_product=False,
    combos=False,
    timeperiod=[15, 30],
)
ema_ind = from_talib(
    func_name='ema',
    price_data=prices,
    cart_product=False,
    combos=False,
    timeperiod=[100, 200, 300],
)
ema_2_ind = from_talib(
    func_name='ema',
    price_data=prices,
    cart_product=False,
    combos=False,
    timeperiod=[600, 800, 1000],
)
rsi_ema_ind = from_talib(
    func_name='ema',
    indicator_data=rsi_ind,
    cart_product=False,
    combos=False,
    timeperiod=[100, 200, 300],
)
ema_is_above = is_above(
    want_to_evaluate=ema_ind,
    price_data=prices,
    cand_ohlc='close'
)
ema_is_below = is_below(
    want_to_evaluate=ema_ind,
    price_data=prices,
    cand_ohlc='close'
)
ema_2_is_above = is_above(
    want_to_evaluate=ema_2_ind,
    price_data=prices,
    cand_ohlc='close'
)
ema_2_is_below = is_below(
    want_to_evaluate=ema_2_ind,
    price_data=prices,
    cand_ohlc='close'
)
rsi_is_below = is_below(
    want_to_evaluate=rsi_ind,
    user_args=[50,60],
)
rsi_ema_is_below = is_below(
    want_to_evaluate=rsi_ema_ind,
    indicator_data=rsi_ind,
)
```

```python
combine_evals(
    first_eval_data=rsi_is_below,
    second_eval_data=rsi_ema_is_below,
    # plot_results=True,
    # first_eval_data_needs_prices=False,
    # second_eval_data_needs_prices=False,
    # first_ind_data=rsi_ind,
    # second_ind_data=rsi_ema_ind,
)
```

```python
combine_evals(
    first_eval_data=ema_is_below,
    second_eval_data=rsi_is_below,
    plot_results=True,
    first_eval_data_needs_prices=True,
    second_eval_data_needs_prices=False,
    first_ind_data=ema_ind,
    second_ind_data=rsi_ind,
    prices=prices,
)
```

```python
combine_evals(
    first_eval_data=ema_is_above,
    second_eval_data=ema_2_is_above,
    plot_results=True,
    first_eval_data_needs_prices=True,
    second_eval_data_needs_prices=True,
    first_ind_data=ema_ind,
    second_ind_data=ema_2_ind,
    prices=prices
)
```
